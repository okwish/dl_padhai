### Numpy

```python
import numpy as np
```

```python
l = [1,2,3]
np.array(l,float) #ndarray
np.asarray(l,float)
#both similar

np.arange(10) #like range - ndarray
```

```python
shape_tup = (2,3)
np.zeros(shape_tup) # There are other arguments also in the np.zeros fn definition
#shape - can be specified with list, .. also
```

```python
np.random.random([4,5]) #normally distributed random numbers
```

```python
X = np.random.random([4,5])
X.T # transpose
X.reshape(20,1) 
```

```python
a = np.a range(5)
a + 1 # broadcasting
# operation broadcasted to each element in nd array
```

numpy math operations are generally applicable to matrices(not scalars alone)  
numpy functions are defined such that they **BROADCAST**  

```python
u = [1,2,3]
v = [-1,0,1]

np.inner(u,v) #inner product
np.outer(u,v) #outer product

# np.dot()
```

```python
a.sum()
a.sum(axis = 0) #column sums - gives a row with sum of cols
a.sum(axis = 1) #row sums
```

Specifying **'axis'** is there is many functions. Along which axis of the tensor, should the specified operation be done.

### Pandas

```python
import pandas as pd
#package to deal with data

d = pd.read_csv('mobile_cleaned.csv') # return dataframe

d.head() #first 5 rows
```

```python
ax = sns.scatterplot(x='stand_by_time',y='battery_capacity',data = d)

#ax - handler for axis
# with handler we can change stuff later

#return an object (handler) and plots 

#its same as ax in matplotlib
#if we don't assign the returned object to a varible, its
#also returned to the 'out'
```

```python
ax = sns.scatterplot(x='stand_by_time',y='battery_capacity',hue = 'thickness',data = d)
```

```python
ax = sns.displot(d['stand_by_time'])#histogram
```

```python
# two general ways of passing data 
# pass dataframe and headings
# pass column, etc.. directly
```

```python
ax = sns.displot(d['stand_by_time'],kde = False,rug=True)
#kde
#rug shows exact pos
```

```python
ax = sns.displot(d['stand_by_time'],kde = False,rug=True, bins=20)
#bin size
```

```python
ax = sns.boxplot(x='is_liked',y='battery_capacity', data = d)
```

```python
ax = sns.boxplot(x='expandable_memory',y='price',data=d)
```

- from boxplot we get a general distribution over one variable for each of the categories
- can look for trends etc.

```python
uniform_data = np.random.rand(10,12) #10x12 matrix

ax = sns.heatmap(uniform_data, cmap='YlGnBu')
#value from color bar

#useful for visualising performance across serveral parameters, etc.
```
