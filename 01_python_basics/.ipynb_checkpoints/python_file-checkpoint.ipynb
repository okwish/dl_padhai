```python
class class_name:
    pass
    
# pass - placeholder for future code
# nothing happens but avoid getting error when empty code is not allowed

class_obj = class_name() #create object
```

```python
class class_name:
    """docstring - which is a description of the class"""
    #all functions in class take first arg as 'self'
    def __init__(self,arg1,arg2): #initialisation function
        self.var1 = arg1
        self.var3 = 5
        #class variables in __init__
        #self.var = .. #initialise in __init__
    def class_fn(self):
        print(self.var3)
    
    
# making object calls the __init__ fn - "initialising"

class_obj = class_name('check1','check2')

print(class_obj.var1)
class_obj.class_fn() #function called with paranthesis

class_name.__doc__ #to print the docstring
```

```python
class sub_class(class_name): #inheritence
    def __init__(self):
        class_name.__init__(self,'check3','check4')
        
sub_class_obj = sub_class()

#inherit only functions
#not variables

#can call main class in sub class definition
```

```python
sub_class_obj.var1
```

```python
class print_class():
    def __str__(self):
        return("this text is returned on using print with class")

print_class_obj = print_class()
    
print(print_class_obj)
```

```python
import pandas as pd
#package to deal with data

d = pd.read_csv('mobile_cleaned.csv')

d.head() #first 5 rows
```

```python
ax = sns.scatterplot(x='stand_by_time',y='battery_capacity',data = d)

#ax - handler for axis

#return an object (handler) and plots 

#its same as ax in matplotlib
#if we don't assign the returned object to a varible, its
#also returned to the 'out'
```

```python
ax = sns.scatterplot(x='stand_by_time',y='battery_capacity',hue = 'thickness',data = d)
```

```python
ax = sns.distplot(d['stand_by_time'])#histogram
```

```python
# two general ways of passing data 
# pass dataframe and headings
# pass column, etc.. directly
```

```python
ax = sns.distplot(d['stand_by_time'],kde = False,rug=True)
#kde
#rug shows exact pos
```

```python
ax = sns.distplot(d['stand_by_time'],kde = False,rug=True, bins=20)
#bin size
```

```python
ax = sns.boxplot(x='is_liked',y='battery_capacity', data = d)
```

```python
ax = sns.boxplot(x='expandable_memory',y='price',data=d)
```

- from boxplot we get a general distribution over one variable for each of the categories
- can look for trends etc.
