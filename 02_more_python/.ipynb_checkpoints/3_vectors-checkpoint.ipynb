```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.quiver(0,0,3,4) #start coord, end coord
plt.show()

# but below the vector scale and axis scale are not the same
```

```python
plt.quiver(0,0,3,4, scale_units='xy',angles='xy',scale=1, color='r')
#scales.. to make axis scale and vector scale same
plt.quiver(0,0,-3,4, scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-10,10) #axis limits
plt.ylim(-10,10)
plt.show()
```

```python
#function to plot a given list of vectors
#each vector is a list-like(of size 2)
def plot_vectors(vecs):
    for vec in vecs:
        plt.quiver(0,0,vec[0],vec[1], scale_units='xy',angles='xy',scale=1, color='r')
    plt.xlim(-10,10) #axis limits
    plt.ylim(-10,10)
    plt.show()
```

```python
plot_vectors([(3,4),(-3,4)])
```
