compute loss for a given dataset(TOTAL loss(square error))  
find parameters with min loss.  
brute search.(argmin in loss)  
plot loss for each parameter-vector.  

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x,w,b):
    return 1/(1+np.exp(-w*x + b))
```

### Toy data

```python
#data
w_unknown = 0.5
b_unknown = 0.25

X = np.random.random(25) * 20 -10 #for changing range from (0,1) to (-10,10)
#25 datapoints
Y = sigmoid(X, w_unknown, b_unknown)

plt.plot(X,Y,'*') #this is actually doing scatter plot because of the '*'
plt.show()
```

### Loss  

(objective function to be minimised)

```python
#function to find loss
#give data - set of input-ouputs
def calc_loss(X,Y,w_est,b_est): #estimates of parameters
    
    loss = 0
    for x, y in zip(X,Y):
        y_pred = sigmoid(x,w_est,b_est) # the model
        loss += (y - y_pred)**2 # square error
    return loss    
```

"Total loss" - sum of square error, **across all datapoints**.  
By loss we mean total loss.  
Actually mean square error (**MSE**), but 1/N is neglected.  
Accounting all datapoints.

```python
#some w,b range
ww = np.linspace(0,2,100)
bb = np.linspace(-1,1,100)
# calculate losses for each of these w,b pairs

W2_grid,B2_grid = np.meshgrid(ww,bb)

Loss_mat = np.zeros(W2_grid.shape) #loss matrix with same shape as grid

# iterating all grid points in meshgrid
for i in range(W2_grid.shape[0]):
    for j in range(W2_grid.shape[1]):
        Loss_mat[i,j] = calc_loss(X, Y, W2_grid[i,j], B2_grid[i,j])


fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(W2_grid,B2_grid,Loss_mat,cmap = 'viridis') 
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('loss')
plt.show()
#ax.view_init(30,270)        
```

Plotting loss wrt parameters. This is the **loss surface**. - "Total loss" for each parameter-vector.  
We optimize in this loss surface - ie, move and reach minimum in the loss surface. The corresponding parameter-vector is the optimal one(minimum total-loss).

```python
#want w, b for which loss is min.
np.argmin(Loss_mat)
```

but input is 2d  
this is after flattening into a vector(lexicographic)(rows stacked)

#### converting lexicographic index to 2d index

```python
ij = np.argmin(Loss_mat)
i = int( np.floor(ij/Loss_mat.shape[1]) ) #cast to int for indexing
j = int( ij - i*Loss_mat.shape[1] )

print(i,j)

print(W2_grid[i,j],B2_grid[i,j])
```
