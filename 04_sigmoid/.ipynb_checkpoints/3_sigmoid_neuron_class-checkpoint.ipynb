#### function to plot sigmoid-function and datapoints - on a passed axis handler

pass an axis handler also to the function. plot on that.

then we can make subplots and see the evolution of the model function

```python
def plot_model_fn(XX,Y,sn,ax): #data, sigmoid neuron learned so far, axis handler
    # input for plotting:
    X_ax = np.linspace(-10,10,100)
    Y_ax = np.linspace(-10,10,100)
    X2_grid, Y2_grid = np.meshgrid(X_ax,Y_ax)
    # output:
    Z_mat = np.zeros(X2_grid.shape) # same as grid shape
    
    for i in range(Y_ax.size):
        for j in range(X_ax.size):
            inp_X = np.asarray( [X_ax[j],Y_ax[i]] )#input vector
            Z_mat[i,j] = sn.forward(inp_X)
            #calling sigmoid from the class - the current parameters are already in it.
            
    # plotting on the passed axis handler
    ax.contourf(X2_grid,Y2_grid,Z_mat,cmap=my_cmap)
    
    #plot datapoints (data is passed as arguments)
    #XX -> input
    #Y -> ouput - use color  
    
    # plotting on the passed axis handler
    ax.scatter(XX[:,0], XX[:,1], c=Y, cmap = my_cmap)
    ax.plot()
```

```python
# plotting evolution of model-function on sublplots

sigmoid_neuron.fit(XX,Y,1,0.75,True) #once with initialization.
N = 20 # number of time (epochs)

#change figure size to accomodate all plots
# figsize = (width, height)
plt.figure(figsize = (10,N*5) )

for i in range(N):
    # print(sigmoid_neuron.W, sigmoid_neuron.b)
    
    #axis handler
    ax = plt.subplot(N,1,i+1) 
    # total Nx1 subplots. in that index: i+1
    
    plot_model_fn(XX,Y,sigmoid_neuron,ax) #passing handler
    #make a plot handler - pass it to the function - function plots on that.
    
    sigmoid_neuron.fit(XX,Y,1,0.75,False) # then with current parameters
```

two things - there is a **decision boundary**     
and a margin - the **sharpness of transition**.  

first it will try to allign the decision boundary well so that error is less..   
once its alligned only thing it can do to reduce error is to make it sharper.
