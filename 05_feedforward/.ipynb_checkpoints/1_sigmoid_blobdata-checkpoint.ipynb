## Blob data

- using make_blobs function

```python
from sklearn.datasets import make_blobs
# for generating data for classification
```

```python
# generate blob data
data, labels = make_blobs(n_samples=1000, # number of datapoints
                          centers=4, # number of blobs - one blob has one label -> number of classes.
                          n_features=2, # number of inputs
                          random_state=0 # same random_state -> generate same data - for repeatability
                         )

# returns:
# data - input matrix
# labels - output vector

# here, 2D inputs

print(data.shape, labels.shape)
print(set(labels))  # set has only unique values
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors
```

#### plot data

```python
# custom color map
my_cmap = matplotlib.colors.LinearSegmentedColormap.from_list(
    "", ["red", "yellow", "green"]
)
# map smallest to red, largest to green.

# plot input data points with color=label(ouput)
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap=my_cmap)
# c = --- -> take color 'value' from that vector.
plt.show()
```

This is actually 3D data. But we can visualise that on 2D (scatter plot) with color as the 3rd value.

Now combine two classes each into one - so finally we have 2 classes. This way we get binary classes data - which are not linearly seperable. If we generate just 2 classes in first place using make_blob - its possible that they could be linearly seperable.

```python
# get 2 classes out of these 4 classes(combining 2 each)

labels_orig = labels
# new label:
labels = np.mod(labels_orig, 2)  # remainder dividing by 2
# 0,2 -> 0
# 1,3 -> 1

plt.scatter(data[:, 0], data[:, 1], c=labels, cmap=my_cmap)
plt.show()
```

This is our data(for binary classification). And its not linearly seperable.

Split into train, validation. Fit using train data.

validation != test.

test data is purely for evalution - completely unseen data - don't adjust hyperparameters with these.
validation data - unseen while training. but we use it to tune hyperparameters.. eg: loss-with-each-epoch and see its overfitting, etc.. thus tuning #epochs.

```python
from sklearn.model_selection import train_test_split
```

```python
# train, validation

XX_train, XX_val, Y_train, Y_val = train_test_split(
    data, labels, stratify=labels, random_state=0
)
print(XX_train.shape, XX_val.shape)
```

### Classification using Sigmoid Neuron

single sigmoid neuron - model

```python
from sigmoid_neuron import SigmoidNeuron
```

## Model 'class'

1. init fuction - parameters - common attributes being updated/used by the functions
2. some fns that characterise the foward pass - given the inputs, compute the output
    - eg : in sigmoid class - linear fn(aggregation), sigmoid fn(activation)
3. then some functions for gradient computation - to be used in fit()
4. fit function (training) - fit the "right model" - GO THROUGH THE DATA - update parameters 
    - several epochs, .. things to display losses after each epoch, etc.. here
5. predit function - given a 'set' of inputs (data)- do forward pass on each input(with 'current' weights) - return a prediction vector. - typically to be done for trained weights.


The abstraction. - like the visual model itself - with functionlities to do stuff, gets trained with data, calling parameters gives the current learned parameters.

```python
sn = SigmoidNeuron()

#call fit to fit. (Train parameters from data(train data))
sn.fit(XX_train, Y_train, epochs=1000, learning_rate=0.5, display_loss=True)
```

adjust #epochs, learning rate above - until a satisfactory loss curve is obtained.   

randomness in initialisation of model-parameters. ouput/optimization will depend on that.

```python
from sklearn.metrics import accuracy_score
```

```python
# evaluation:

Y_pred_train = sn.predict(XX_train) 

# above will be sigmoid output. value b/w 0 and 1.
# binarise for class ouput (threshold)
Y_pred_binarised_train = (Y_pred_train >= 0.5).astype("int").ravel()

Y_pred_val = sn.predict(XX_val)
Y_pred_binarised_val = (Y_pred_val >= 0.5).astype("int").ravel()# class output

accuracy_train = accuracy_score(Y_pred_binarised_train, Y_train)
accuracy_val = accuracy_score(Y_pred_binarised_val, Y_val)

print("Training accuracy", round(accuracy_train, 2)) #round value to 2 decimal places.
print("Validation accuracy", round(accuracy_val, 2))
```

baseline is 0.5 . if it were predicting one class all the time - it will get ~0.5 accuracy.

0.7 is not too bad.. but not good either.

### Visualizing prediction

using 2 indicators - color, size  

color = predicted label  
size = if the prediction correct(say, small if correct, else larger)

visualise:  
datapoints  
color = predicted label  
size = 15*abs(prediction-true) + 0.2  
    = 0.2 if right prediction
    = 15.2 if wrong prediction

y_pred_bin - y_train -> 0 if right (same); 1 or -1 if wrong(diff) . (abs)

4 things in the plot - x,y,color,size

```python
plt.scatter(XX_train[:,0], XX_train[:,1], c=Y_pred_binarised_train, cmap=my_cmap, s=15*(np.abs(Y_pred_binarised_train-Y_train)+.2))
#color = predicted label
#small => right prediction
plt.show()
```

predicting points to either side of a 'line'  
ie, points on one side is predicted as one class; and other side points are predicted as the other class.

=> single sigmoid neuron is trying to seperate them linearly.
