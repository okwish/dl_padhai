```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors
```

```python
np.random.seed(0) # singleton RandomState
```

```python
from sklearn.metrics import accuracy_score
```

```python
# custom color map
my_cmap = matplotlib.colors.LinearSegmentedColormap.from_list(
    "", ["red", "yellow", "green"]
)
```

### Data (blob)

```python
from sklearn.datasets import make_blobs

# generate blob data
data, labels = make_blobs(n_samples=1000, # number of datapoints
                          centers=4, # number of blobs - one blob has one label -> number of classes.
                          n_features=2, # number of inputs
                          random_state=0 # same random_state -> generate same data - for repeatability
                         )

# need multi-class data
# originally we made 4 classes
# don't merge them like before

from sklearn.model_selection import train_test_split

XX_train, XX_val, Y_train, Y_val = train_test_split(
    data, labels, stratify=labels, random_state=0
)
```

### One hot encoding

current y - labels (single values) 
but we want a probability distribution.  
vector with corresponding index=1, rest=0 -> one hot

0 -> (1, 0, 0, 0), 1 -> (0, 1, 0, 0), 2 -> (0, 0, 1, 0), 3 -> (0, 0, 0, 1)

convert categorical value into one hot vectors.

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
# 0 -> (1, 0, 0, 0), 1 -> (0, 1, 0, 0), 2 -> (0, 0, 1, 0), 3 -> (0, 0, 0, 1)
YY_OH_train = enc.fit_transform(np.expand_dims(Y_train,1)).toarray()
YY_OH_val = enc.fit_transform(np.expand_dims(Y_val,1)).toarray()
print(YY_OH_train.shape, YY_OH_val.shape)


# np.expand_dims( arr, axis=() )
# change "dimension" of the input array - 1 in the specified axes in the dimension

# eg: dim = m; axis=(0,1) becomes 1x1xm , ie 1 in axis 0 and axis 1

# above dim = m; axis = 1 => become mx1 -> makes a column - because that is needed for fit_transform()
# before we used reshape(-1,1) for the same purpose (ie, to make it into column vector)
```


##  FF net with multiple-output for multiclass classification

previous - one ouput

#### Softmax
can use sigmoid only for binary class ouput. We need a probability distribution across the classes - softmax activation in the output layer.

#### Cross entropy loss
ground truth - one hot  
model output - prob distribution (softmax)  

compare two prob distributions using cross entropy loss  
a number quantifying the difference between two distributions(vectors)

```python
# FF model class with multiple output(softmax)
from multiclass_ff_net import MultiClassFFNet
```

```python
mfn = MultiClassFFNet(2,4,[2,3])

mfn.fit(XX_train,YY_OH_train,epochs=2000,learning_rate=.005,display_loss=True)
```

oscillations => lower the learning rate.

network ouput is prob vector. but we need the single-class-value as inference.  

=> class corresponding to max prob. -> argmax 

use this for finding accuracy.

```python
# evaluation

YY_pred_train = mfn.predict(XX_train)
Y_pred_train = np.argmax(Y_pred_train,1)

YY_pred_val = mfn.predict(XX_val)
Y_pred_val = np.argmax(Y_pred_val,1)

accuracy_train = accuracy_score(Y_pred_train, Y_train)
accuracy_val = accuracy_score(Y_pred_val, Y_val)

print("Training accuracy", round(accuracy_train, 2))
print("Validation accuracy", round(accuracy_val, 2))
```

```python
plt.scatter(XX_train[:,0], XX_train[:,1], 
            c=Y_pred_train, cmap=my_cmap, 
            s=15*(np.abs(np.sign(Y_pred_train-Y_train))+.1))
plt.show()
```
