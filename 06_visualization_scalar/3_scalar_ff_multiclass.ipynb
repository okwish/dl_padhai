```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors
```

```python
# custom color map
my_cmap = matplotlib.colors.LinearSegmentedColormap.from_list(
    "", ["red", "yellow", "green"]
)
```

#### Data (blob)

```python
from sklearn.datasets import make_blobs

# generate blob data
data, labels = make_blobs(n_samples=1000, # number of datapoints
                          centers=4, # number of blobs - one blob has one label -> number of classes.
                          n_features=2, # number of inputs
                          random_state=0 # same random_state -> generate same data - for repeatability
                         )

from sklearn.model_selection import train_test_split

XX_train, XX_val, Y_train, Y_val = train_test_split(
    data, labels, stratify=labels, random_state=0
)
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
# 0 -> (1, 0, 0, 0), 1 -> (0, 1, 0, 0), 2 -> (0, 0, 1, 0), 3 -> (0, 0, 0, 1)
YY_OH_train = enc.fit_transform(np.expand_dims(Y_train,1)).toarray()
YY_OH_val = enc.fit_transform(np.expand_dims(Y_val,1)).toarray()
print(YY_OH_train.shape, YY_OH_val.shape)
```

### Multiclass FF net - scalar implementation

This is not the generic class like before.   
Done this way to implement 'scalar' backpropagation - and to visualize weights, hidden neuron functions, ..

<img src="model_pictures/multiple_out_ff_net.png" width="40%" height="40%" />

```python
from multiclass_ff_scalar import FF_MultiClass_Scalar
```

```python
mfs = FF_MultiClass_Scalar()
```

```python
weight_matrices = []
mfs.fit(XX_train,YY_OH_train,weight_matrices,
        epochs=2000,learning_rate=1,
        display_loss=True, display_weight=True)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# evaluation

YY_pred_train = mfs.predict(XX_train)
Y_pred_train = np.argmax(YY_pred_train,1)

YY_pred_val = mfs.predict(XX_val)
Y_pred_val = np.argmax(YY_pred_val,1)

accuracy_train = accuracy_score(Y_pred_train, Y_train)
accuracy_val = accuracy_score(Y_pred_val, Y_val)

print("Training accuracy", round(accuracy_train, 2))
print("Validation accuracy", round(accuracy_val, 2))


plt.scatter(XX_train[:,0], XX_train[:,1], 
            c=Y_pred_train, cmap=my_cmap, 
            s=15*(np.abs(np.sign(Y_pred_train-Y_train))+.1))
plt.show()
```
