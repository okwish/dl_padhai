```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Data (blob)

```python
from sklearn.datasets import make_blobs

# generate blob data
data, labels = make_blobs(
    n_samples=1000,  # number of datapoints
    centers=4,  # number of blobs - one blob has one label -> number of classes.
    n_features=2,  # number of inputs
    random_state=0,  # same random_state -> generate same data - for repeatability
)

from sklearn.model_selection import train_test_split

XX_train, XX_val, Y_train, Y_val = train_test_split(
    data, labels, stratify=labels, random_state=0
)
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
# 0 -> (1, 0, 0, 0), 1 -> (0, 1, 0, 0), 2 -> (0, 0, 1, 0), 3 -> (0, 0, 0, 1)
YY_OH_train = enc.fit_transform(np.expand_dims(Y_train, 1)).toarray()
YY_OH_val = enc.fit_transform(np.expand_dims(Y_val, 1)).toarray()
print(YY_OH_train.shape, YY_OH_val.shape)
```

#### plotting gradient. 

gradiant $\propto$ ( $w_t$ - $w_{t-1}$  ), ie change/update in parameters  
by taking update in parameters - we are getting gradient-ish  

gradient plot is some kind of window to the loss surface.  
also how much the parameters are getting updated, etc... 

```python
import matplotlib.colors

# custom color map
my_cmap = matplotlib.colors.LinearSegmentedColormap.from_list(
    "", ["red", "yellow", "green"]
)

from sklearn.metrics import accuracy_score



# 1. training, validation accuracy
# 2. prediction plot of datapoints
# 3. plot gradients of each parameters (at each epoch)
# find gradient using (wt - wt-1)

def evaluations(plot_scatter=False, plot_gradient=True, plot_scale=0.1):
    
    # 1. training and validation accuracy and print it.
    YY_pred_train = model.predict(XX_train)
    Y_pred_train = np.argmax(YY_pred_train, 1)
    
    YY_pred_val = model.predict(XX_val)
    Y_pred_val = np.argmax(YY_pred_val, 1)
    
    accuracy_train = accuracy_score(Y_pred_train, Y_train)
    accuracy_val = accuracy_score(Y_pred_val, Y_val)
    print("Training accuracy", round(accuracy_train, 4))
    print("Validation accuracy", round(accuracy_val, 4))

    
    # 2. plot data points
    # color = prediction
    # size = right prediction or not. bigger if wrong prediction
    if plot_scatter:
        
        plt.scatter(
            XX_train[:, 0], XX_train[:, 1],
            c=Y_pred_train,
            cmap=my_cmap,
            s=15 * (np.abs(np.sign(Y_pred_train - Y_train)) + 0.1),
        )
        plt.show()

    # 3. PLOT GRADIENTS
    # look at all parameters and see the changes in them with each epoch.
    # plotting the update. w(t) - w(t-1), as that is gradient-ish.
    if plot_gradient:
        # parameter - history values from model
        HH = np.asarray(model.params_h) # params_h is a list
        HH_diff = HH[0:-1, :] - HH[1:, :]  # difference => 'updates'
        # first one 0 to last-1, second one 1 to last. 
        # now those elements are alligned; take diff.

        # subplot
        for i in range(18):  # all 18 parameters in this network.
            plt.subplot(6, 3, i + 1)  # index start with 1(not 0)
            plt.plot(HH_diff[:, i], "-")
            plt.ylim((-plot_scale, plot_scale)) #y-range
            plt.yticks([])  # empty list. so no markings.. to avoid crowding.
            plt.xticks([])
        plt.show()
```

```python
#initialise the mode
#fit the model
#call evaluations
```

#### Test run

```python
model = FFNet(init_method='xavier', activation_function='sigmoid')
model.fit(XX_train, YY_OH_train, epochs=10, eta=1, algo="GD", display_loss=True)
evaluations()
```

### Trying multiple experiments with loops

typically have things like this.. to try multiple experiments  

trying all combinations of initialisation-activation

```python
# different initialisations for sigmoid activation
for init_method in ["zeros", "random", "xavier", "he"]:
    for activation_function in ["sigmoid"]:
        print(init_method, activation_function)
        model = FFNet(init_method=init_method, activation_function=activation_function)
        model.fit(XX_train, YY_OH_train, epochs=50, eta=1, algo="GD", display_loss=True)
        evaluations(plot_scale=0.05)
        print("\n--\n")
```

For sigmoid, xavier working well. but 'He' not doing much.

```python
# different initialisations for tanh activation
for init_method in ["zeros", "random", "xavier", "he"]:
    for activation_function in ["tanh"]:
        print(init_method, activation_function)
        model = FFNet(init_method=init_method, activation_function=activation_function)
        model.fit(XX_train, YY_OH_train, epochs=100, eta=0.5, algo="NAG", display_loss=True)
        evaluations(plot_scale=0.05)
        print("\n--\n")
```

0 initialisation - not update at all..   
as value of tanh at 0 is 0. all activations,.. remain 0.  
small values - only numerical error, etc..   

tanh - more oscillatory behaviour(gradient plots are oscillatory) - because of how the surface is(narrow valley)..  

```python
# different initialisations for relu activation
for init_method in ["zeros", "random", "xavier", "he"]:
    for activation_function in ["relu"]:
        print(init_method, activation_function)
        model = FFNet(init_method=init_method, activation_function=activation_function)
        model.fit(XX_train, YY_OH_train, epochs=50, eta=0.25, algo="GD", display_loss=True)
        evaluations(plot_scale=0.05)
        print("\n--\n")
```

typically its recommended to use small learning rates with relu, as it has derivative=1 for all positive x, give large updates... chance for unstability.

some parameters are not getting updated at all - because of the 0 derivative in negative region.  
(update curve remain const throughout -> that parameter is not getting updated. )

much higher accuracy with a simple loss function. (less computation.)  
computationally more tractable.

```python
# different initialisations for leaky relu activation
for init_method in ["zeros", "random", "xavier", "he"]:
    for activation_function in ["leaky_relu"]:
        print(init_method, activation_function)
        model = FFNet(
            init_method=init_method, activation_function=activation_function, leaky_slope=0.1
        )
        model.fit(XX_train, YY_OH_train, epochs=50, eta=0.5, algo="GD", display_loss=True)
        evaluations(plot_scale=0.05)
        print("\n--\n")
```
