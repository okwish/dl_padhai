```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
import matplotlib.colors

# custom color map
my_cmap = matplotlib.colors.LinearSegmentedColormap.from_list(
    "", ["red", "yellow", "green"]
)
```

### Data (iris)

toy dataset

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
data = iris.data[:, :2]  # take only the first two features (for visualization purpose)
labels = iris.target

plt.scatter(data[:,0], data[:,1], c=labels, cmap=my_cmap)
plt.show()

print("Data shape",data.shape)
print("Labels shape",labels.shape)

# iris.data.shape = (150,4)
```

```python
set(labels) # 3 classes
```

```python
from sklearn.model_selection import train_test_split


XX_train, XX_val, Y_train, Y_val = train_test_split(
    data, labels, stratify=labels, random_state=0
)

from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
# 0 -> (1, 0, 0, 0), 1 -> (0, 1, 0, 0), 2 -> (0, 0, 1, 0), 3 -> (0, 0, 0, 1)
YY_OH_train = enc.fit_transform(np.expand_dims(Y_train, 1)).toarray()
YY_OH_val = enc.fit_transform(np.expand_dims(Y_val, 1)).toarray()
print(YY_OH_train.shape, YY_OH_val.shape)
```

### FF class  

L2 regularization  - additional L2 norm of weights in loss function (objective)  

code part:  in fit - where gradient is found - additional term

also, tracking both training and validation error  

training and validation error - checking overfitting, .. 

```python
# function to find accuracy after training

from sklearn.metrics import accuracy_score

def print_accuracy():
    YY_pred_train = model.predict(XX_train)
    Y_pred_train = np.argmax(YY_pred_train, 1)
    
    YY_pred_val = model.predict(XX_val)
    Y_pred_val = np.argmax(YY_pred_val, 1)
    
    accuracy_train = accuracy_score(Y_pred_train, Y_train)
    accuracy_val = accuracy_score(Y_pred_val, Y_val)
    
    print("Training accuracy", round(accuracy_train, 4))
    print("Validation accuracy", round(accuracy_val, 4))
```

More and more complex models can help reduce bias - but may lead to overfitting.  
there we can use regularization to avoid overfitting.

bias - gap from a good accuracy(eg: 0.99 marked here)

gap b/w training,validation - variance

### Changing model complexity
number of hidden layer; number of parameters

```python
# validation data also passed to find validation accuracy after each epoch
```

```python
model = FFNet(num_hidden=1)
model.fit(XX_train, YY_OH_train, 
          XX_val, Y_val,           
          epochs=100, eta=0.5, 
          l2_norm=False, display_loss=False)
print_accuracy()
```

```python
model = FFNet(num_hidden=2)
model.fit(XX_train, YY_OH_train, 
          XX_val, Y_val,           
          epochs=100, eta=1, 
          l2_norm=False, display_loss=False)
print_accuracy()
```

```python
model = FFNet(num_hidden=4)
model.fit(XX_train, YY_OH_train, 
          XX_val, Y_val,           
          epochs=400, eta=0.25, 
          l2_norm=False, display_loss=False)
print_accuracy()
```

```python
model = FFNet(num_hidden=8)
model.fit(XX_train, YY_OH_train, 
          XX_val, Y_val,           
          epochs=500, eta=0.2, 
          l2_norm=False, display_loss=False)
print_accuracy()
```

```python
model = FFNet(num_hidden=32)
model.fit(XX_train, YY_OH_train, 
          XX_val, Y_val,           
          epochs=500, eta=0.2, 
          l2_norm=False, display_loss=False)
print_accuracy()
```

```python
model = FFNet(num_hidden=64)
model.fit(XX_train, YY_OH_train, 
          XX_val, Y_val,           
          epochs=2000, eta=0.2, 
          l2_norm=False, display_loss=False)
print_accuracy()
```

training accuracy improving, but not validation accuracy. overfitting.
