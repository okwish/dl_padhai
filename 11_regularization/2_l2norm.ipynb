#### Data (iris)

toy dataset

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
data = iris.data[:, :2]  # take only the first two features (for visualization purpose)
labels = iris.target

print("Data shape",data.shape)
print("Labels shape",labels.shape)

# iris.data.shape = (150,4)
```

```python
from sklearn.model_selection import train_test_split


XX_train, XX_val, Y_train, Y_val = train_test_split(
    data, labels, stratify=labels, random_state=0
)

from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
# 0 -> (1, 0, 0, 0), 1 -> (0, 1, 0, 0), 2 -> (0, 0, 1, 0), 3 -> (0, 0, 0, 1)
YY_OH_train = enc.fit_transform(np.expand_dims(Y_train, 1)).toarray()
YY_OH_val = enc.fit_transform(np.expand_dims(Y_val, 1)).toarray()
print(YY_OH_train.shape, YY_OH_val.shape)
```

### FF class  

L2 regularization  - additional L2 norm of weights in loss function (objective)  

code part:  in fit - where gradient is found - additional term

also, tracking both training and validation error  

training and validation error - checking overfitting, .. 

```python
# function to find accuracy after training

from sklearn.metrics import accuracy_score

def print_accuracy():
    YY_pred_train = model.predict(XX_train)
    Y_pred_train = np.argmax(YY_pred_train, 1)
    
    YY_pred_val = model.predict(XX_val)
    Y_pred_val = np.argmax(YY_pred_val, 1)
    
    accuracy_train = accuracy_score(Y_pred_train, Y_train)
    accuracy_val = accuracy_score(Y_pred_val, Y_val)
    
    print("Training accuracy", round(accuracy_train, 4))
    print("Validation accuracy", round(accuracy_val, 4))
```

### L2 regularization

```python
model = FFNet(num_hidden=64)
model.fit(XX_train, YY_OH_train, 
          XX_val, Y_val,           
          epochs=2000, eta=0.1, 
          l2_norm=True, lambda_val=0.1,
          display_loss=True
         )
print_accuracy()
```

loss and norm sq of parameter tensor.

```python
model = FFNet(num_hidden=64)
model.fit(XX_train, YY_OH_train, 
          XX_val, Y_val,           
          epochs=2000, eta=0.1, 
          l2_norm=True, lambda_val=1,
          display_loss=True
         )
print_accuracy()
```

see how they are changing -
