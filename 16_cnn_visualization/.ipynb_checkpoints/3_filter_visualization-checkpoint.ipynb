64 filters - each 11 x 11 x 3

**weight tensor** - at a convolution layer.   
There will be k filters of dimension rows x cols x depth.

Can't call numpy() on Tensor that requires grad. Use tensor.detach().numpy() instead.   

(tensor.data is an old field. Not to be used anymore.)

```python
def plot_filters_each_channel(weight_tensor):
    
    # plot each slice in each kernel as a grey-scale image
    
    kernel_width = weight_tensor.shape[2]
    kernel_height = weight_tensor.shape[3]
    
    n_slices = weight_tensor.shape[1]    
    n_kernels = weight_tensor.shape[0]    
    
    n_plots = n_kernels * n_slices
    
    plot_cols = 12
    plot_rows = n_plots//plot_cols + 1
    
    npimg = np.array(weight_tensor.detach().numpy(), np.float32)
    
    axis_index = 0
    
    fig = plt.figure( figsize=(plot_cols, plot_rows) )    
    
    for i in range(n_kernels):
        for j in range(n_slices):
            axis_index += 1
            ax = fig.add_subplot(plot_rows, plot_cols, axis_index)
            
            # j-th depth slice in i-th kernel 
            npimg = np.array(weight_tensor[i, j].detach().numpy(), np.float32)
            
            npimg = (npimg - np.mean(npimg)) / np.std(npimg)
            npimg = np.minimum(1, np.maximum(0, (npimg + 0.5)))
            
            ax.imshow(npimg)
            ax.set_title(str(i) + ',' + str(j)) # i-th kernel; j-th slice in it
            ax.axis('off')
            ax.set_xticklabels([])
            ax.set_yticklabels([])
   
    plt.tight_layout()
    plt.show()
```

```python
def plot_filters_as_rgb(weight_tensor):
    
    # plot each kernel as an rgb image
    # only possible if kernel depth is 3
    
    num_kernels = weight_tensor.shape[0]    
    
    plot_cols = 12
    plot_rows = num_kernels//plot_cols + 1
    
    fig = plt.figure(figsize=(plot_cols,plot_rows))
    
    for i in range(num_kernels):
    
        ax = fig.add_subplot(plot_rows,plot_cols,i+1)
        
        # weight_tensor[i] => i-th kernel (depth=3) => RGB image
        npimg = np.array(weight_tensor[i].detach().numpy(), np.float32)
        
        # normalise, transpose to plot
        npimg = (npimg - np.mean(npimg)) / np.std(npimg)
        npimg = np.minimum(1, np.maximum(0, (npimg + 0.5)))
        npimg = npimg.transpose((1, 2, 0))
        
        ax.imshow(npimg)
        
        ax.axis('off')
        ax.set_title(str(i)) # kernel number
        ax.set_xticklabels([])
        ax.set_yticklabels([])

    plt.tight_layout()
    plt.show()
```

```python
def plot_weights_collated(weight_tensor):
    
    # plot all the weights in a single plot
    # each row is a kernel 
    # one row - slices of the kernel stacked(horizontally) one after other 
    
    kernel_width = weight_tensor.shape[2]
    kernel_height = weight_tensor.shape[3]
    
    n_slices = weight_tensor.shape[1] # depth of each kernel
    # number of depth slices in each kernel
    
    n_kernels = weight_tensor.shape[0]    
    
    plot_rows = n_kernels * kernel_width
    plot_cols = n_slices * kernel_height
          
    npimg = np.array(weight_tensor.detach().numpy(), np.float32)
    
    # not very clear about below
    
    npimg = npimg.transpose((0, 2, 1, 3)) 
    # num_kernels x kernel_width x n_slices x kernel_height
    
    # all the weights
    # flatten(ravel) - and reshape into size we are plotting. 
    
    npimg = npimg.ravel().reshape(plot_rows, plot_cols)
    
    npimg = npimg.T # (?)
    
    fig, ax = plt.subplots( figsize=(plot_cols/10, plot_rows/200) )    
    
    # heatmap
    hm_plot = sns.heatmap(npimg, xticklabels=False, yticklabels=False, cmap='Greys', ax=ax, cbar=False)
```

```python
def plot_weights(model, layer_num, each_channel = True, collated = False):
    
    # calling the above functions
    
    layer = model.features[layer_num]
    # n-th in the sequantial(list) - "features"
    
    # check if its a convoln layer.
    # isinstance - check if an object is of a particular object-type(class)
    if isinstance(layer, nn.Conv2d):         
        
        # model.features[].weight - will access the weight values . lly bias values.
        # the attributes of that layer are named that way.
        weight_tensor = model.features[layer_num].weight
        
        if each_channel:
            if collated:
                plot_weights_collated(weight_tensor)
            else:
                plot_filters_each_channel(weight_tensor)
        else:
            if weight_tensor.shape[1] == 3: # only when 3 channel.
                plot_filters_as_rgb(weight_tensor)
            else:
                'Kernels are not 3-channel!! Set each_channel to True.'
    else:
        print('Can only visualise layers which are convolutional.')
```
