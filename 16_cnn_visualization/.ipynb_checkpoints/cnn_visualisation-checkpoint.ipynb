# Utils

### Visualise image(input)

```python
def imshow(img, title):
    
    # inverse correction for normalisation
    std_correction = np.asarray([0.229, 0.224, 0.225]).reshape(3, 1, 1)
    mean_correction = np.asarray([0.485, 0.456, 0.406]).reshape(3, 1, 1)
    npimg = np.multiply(img.numpy(), std_correction) + mean_correction #np.multiply - braodcast
    
    plt.figure(figsize=(batch_size * 4, 4))
    plt.axis('off')
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.title(title)
    plt.show()
```

```python
def show_batch_images(dataloader):
    images, _ = next(iter(dataloader))
    
    outputs = model(images) #prediction
    _, pred = torch.max(outputs.data, 1)
        
    img = torchvision.utils.make_grid(images)
    imshow(img, title=[classes[x.item()] for x in pred])
    
    return images, pred
```

```python
images, pred = show_batch_images(dataloader)
```

### Occlusion analysis

```python
outputs = model(images)
print(outputs.shape)

outputs = nn.functional.softmax(outputs, dim=1) # applying softmax fn on the ouput of model. 
# as no softmax in the model part. it comes in the loss part.

prob_no_occ, pred = torch.max(outputs.data, 1)
print(prob_no_occ)
prob_no_occ = prob_no_occ[0].item()
print(prob_no_occ) # prob predicted when no occlusion.
```

```python
# change input image - pass thr' model - get prob.

def occlusion(model, image, label, occ_size=50, occ_stride=50, occ_pixel=0.5):
    
    # size of occlusion
    # stride (such that no overlap)
    # value of color replaced in the location
    
    width, height = image.shape[-2], image.shape[-1] #using negative indices. this works both when there is batching and no batching.
    
    output_height = int(np.ceil((height-occ_size)/occ_stride))
    output_width = int(np.ceil((width-occ_size)/occ_stride))
    
    heatmap = torch.zeros((output_height, output_width))
    
    for h in range(0, height):
        for w in range(0, width):
            
            h_start = h*occ_stride
            w_start = w*occ_stride
            h_end = min(height, h_start + occ_size)
            w_end = min(width, w_start + occ_size)
            
            if (w_end) >= width or (h_end) >= height:
                continue
            
            input_image = image.clone().detach() # clone => deep copy
            input_image[:, :, w_start:w_end, h_start:h_end] = occ_pixel # occlude
            
            output = model(input_image)
            output = nn.functional.softmax(output, dim=1) # softmax of ouput
            prob = output.tolist()[0][label] # not max. prob corresponding to the one that we are seeing.
            
            heatmap[h, w] = prob # plotting the prob itself. (in theory we took diff from original) 

    return heatmap
```

```python
# printing prob output - softmax value. (prob.)
```

```python
heatmap = occlusion(model, images, pred[0].item(), 32, 14)
```

```python
imgplot = sns.heatmap(heatmap, xticklabels=False, yticklabels=False, vmax=prob_no_occ)
# setting max as prob when no occlusion.
```

11x11 is fairly large kernel size. this gives us a chance to see what they are studying in the initial layer.

64 11x11x3 kernels.  
one way - since kernel depth = 3 here (also they are acting on RGB.) - plot 64 11x11 RGB kernel-images.  
(this is possible only for input layer. (3 channels) )

or each kernel independently .. 3*64 11x11 slices.

```python
def plot_filters_single_channel_big(t):
    
    nrows = t.shape[0]*t.shape[2]
    ncols = t.shape[1]*t.shape[3]
          
    npimg = np.array(t.numpy(), np.float32)
    npimg = npimg.transpose((0, 2, 1, 3))
    
    # all the weights ..
    # flatten(ravel) - and reshape into size we are plotting. 
    
    npimg = npimg.ravel().reshape(nrows, ncols)
    
    npimg = npimg.T
    
    fig, ax = plt.subplots(figsize=(ncols/10, nrows/200))    
    
    # heatmap
    imgplot = sns.heatmap(npimg, xticklabels=False, yticklabels=False, cmap='Greys', ax=ax, cbar=False)
```

```python
def plot_filters_single_channel(t):
    
    nplots = t.shape[0]*t.shape[1]
    ncols = 12
    nrows = 1 + nplots//ncols
    
    npimg = np.array(t.numpy(), np.float32)
    
    count = 0
    
    fig = plt.figure(figsize=(ncols, nrows))
    for i in range(t.shape[0]):
        for j in range(t.shape[1]):
            count += 1
            ax1 = fig.add_subplot(nrows, ncols, count)
            npimg = np.array(t[i, j].numpy(), np.float32)
            npimg = (npimg - np.mean(npimg)) / np.std(npimg)
            npimg = np.minimum(1, np.maximum(0, (npimg + 0.5)))
            ax1.imshow(npimg)
            ax1.set_title(str(i) + ',' + str(j))
            ax1.axis('off')
            ax1.set_xticklabels([])
            ax1.set_yticklabels([])
   
    plt.tight_layout()
    plt.show()
```

```python
def plot_filters_multi_channel(t):
    
    num_kernels = t.shape[0]    
    
    num_cols = 12
    num_rows = num_kernels
    
    fig = plt.figure(figsize=(num_cols,num_rows))
    for i in range(t.shape[0]):
        
        ax1 = fig.add_subplot(num_rows,num_cols,i+1)
        
        # t[i] -> RGB image
        npimg = np.array(t[i].numpy(), np.float32)
        npimg = (npimg - np.mean(npimg)) / np.std(npimg)
        npimg = np.minimum(1, np.maximum(0, (npimg + 0.5)))
        npimg = npimg.transpose((1, 2, 0))
        ax1.imshow(npimg)
        
        ax1.axis('off')
        ax1.set_title(str(i))
        ax1.set_xticklabels([])
        ax1.set_yticklabels([])

    plt.tight_layout()
    plt.show()
```

```python
def plot_weights(model, layer_num, single_channel = True, collated = False):
    
    layer = model.features[layer_num]
    
    if isinstance(layer, nn.Conv2d): # check if its a convoln layer.
        
        # model.features[].weight - will access the weight values . lly bias values.
        # .data gives the value
        weight_tensor = model.features[layer_num].weight.data
        
        if single_channel:
            if collated:
                plot_filters_single_channel_big(weight_tensor)
            else:
                plot_filters_single_channel(weight_tensor)
        else:
            if weight_tensor.shape[1] == 3: # only when 3 channel.
                plot_filters_multi_channel(weight_tensor)
            else:
                'Can only plot weights with three channels with single_channel = False'
    else:
        print('Can only visualise layers which are convolutional')
```
